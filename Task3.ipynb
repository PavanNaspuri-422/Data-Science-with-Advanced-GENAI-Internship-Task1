**1.Maximum Product of Two Elements in an Array**
Example 1:
Input: nums = [3,4,5,2]
Output: 12

```python
nums=list(map(int,input().split(",")))
res = 0
for i in range(len(nums)):
    for j in range(len(nums)):
        if i != j:
            res = max(res, (nums[i] - 1) * (nums[j] - 1))
            
print(res)
```

**2.Count Number of Teams**
Example 1:
Input: rating = [2,5,3,4,1]
Output: 3

```python
rating=list(map(int,input().split(",")))
n = len(rating)
count = 0
for i in range(1, n-1):
    left_smaller = right_larger = left_larger = right_smaller = 0
    for j in range(i):
        if rating[j] < rating[i]:
            left_smaller += 1
        elif rating[j] > rating[i]:
            left_larger += 1
    for k in range(j + 1, n):
        if rating[k] > rating[i]:
            right_larger += 1
        elif rating[k] < rating[i]:
            right_smaller += 1
            
    count += left_smaller * right_larger + left_larger * right_smaller
          
print(count)
```

**3.Number of Students Doing Homework at a Given Time**
Example 1:
Input: startTime = [1,2,3], endTime = [3,2,7], queryTime = 4
Output: 1

```python
startTime=list(map(int,input().split(",")))
endTime=list(map(int,input().split(",")))
queryTime=int(input())
j=0
count=0
for i in startTime:
    if queryTime >= i and endTime[j] >= queryTime:
        count=count+1
    j=j+1
        
print(count)
```

**4.Number of Steps to Reduce a Number to Zero**
Example 1:
Input: num = 14
Output: 6

```python
num=int(input())
count=0
while num!=0:
    count+=1
    if num%2==0:
        num = num/2
    else:
        num-=1
        
print(count)
```

**5.Counting Bits**
Example 1:
Input: n = 2
Output: [0,1,1]

```python
n=int(input())
dp=[0]*(n+1)
sub=1
for i in range(1,n+1):
    if sub*2==i:
        sub=i
    dp[i]=dp[i-sub]+1
    
print(dp)
        
```
